```python
df = pd.read_csv('DataBase\\E20oxides.csv')

available_mat = ['SiO2', 'Al2O3', 'MgO', 'CaO', 'Na2O', 'K2O','ZnO', 'TiO2']
oxides = ['SiO2', 'B2O3', 'Al2O3', 'MgO', 'CaO', 'BaO', 'Li2O', 'Na2O', 'K2O',
       'ZnO', 'SrO', 'PbO', 'Fe2O3', 'Y2O3', 'La2O3', 'TiO2', 'GeO2', 'ZrO2',
       'P2O5', 'V2O5']

df_reduit = df


for i in range(len(oxides)):
    if oxides[i] not in available_mat:
        df_reduit = df_reduit[ df_reduit[oxides[i]] < 0.01 ]
        df_reduit = df_reduit.drop(oxides[i], axis=1)
```
