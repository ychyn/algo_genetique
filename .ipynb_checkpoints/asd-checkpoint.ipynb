```python
def fitness(prop,weight):
    rho,E,Tg,Tm	= prop
    wrho,wE,wTg,wTm	= weight
    return wrho * (1 - rho) + wE * E + wTg * (1 - Tg) + wTm * (1 - Tm)
    
```

```python
weight = [0,0.5,0,0.5]
def fitness_new(prop):
    return fitness(prop, weight)
```

```python
df['F'] = df.iloc[:,-4:].apply(fitness_new,axis=1)
```
